```python
# data_analysis.py

import pandas as pd

data = {
    "Name": ["Aarav", "Diya", "Rohan", "Sneha", "Kiran"],
    "Maths": [88, 92, 79, 95, 67],
    "Science": [90, 85, 80, 89, 70]
}

df = pd.DataFrame(data)
df["Average"] = df[["Maths", "Science"]].mean(axis=1)

print("Student Data:\n", df)
print("\nTop Performer:")
print(df[df["Average"] == df["Average"].max()])
```
